```python
df=pd.read_csv('heart_failure_clinical_records_dataset.csv')
live= df[df["DEATH_EVENT"] == 0]
dead=df[df["DEATH_EVENT"] == 1]
l_smoking=live['smoking'].tolist()
d_smoking=dead['smoking'].tolist()
l_anaemia=live['anaemia'].tolist()
d_anaemia=dead['anaemia'].tolist()
l_hbp=live['high_blood_pressure'].tolist()
d_hbp=dead['high_blood_pressure'].tolist()
df
```

```python

matrix=np.zeros([10,10],float)
for a in range(0,10):
    for b in range(0,10):
        r=0
        q=0
        s=0
        if l_smoking[a]==1 and d_smoking[b]==0:
            r+=1
        if l_anaemia[a]==1 and d_anaemia[b]==0:
            r+=1
        if l_hbp[a]==1 and d_hbp[b]==0:
            r+=1
        
        if l_smoking[a]==0 and d_smoking[b]==1:
            s+=1
        if l_anaemia[a]==0 and d_anaemia[b]==1:
            s+=1
        if l_hbp[a]==0 and d_hbp[b]==1:
            s+=1
        
        if l_smoking[a]==1 and d_smoking[b]==1:
            q+=1
        if l_anaemia[a]==1 and d_anaemia[b]==1:
            q+=1
        if l_hbp[a]==1 and d_hbp[b]==1:
            q+=1  
        try:    
          matrix[a,b]=(r+s)/(q+r+s)     
        except:
            matrix[a,b]=0
matrix
```

```python
plt.imshow(matrix)
plt.colorbar()
plt.show()
```

```python
xlm = pd.read_csv('Binance_XLMUSDT_d.csv')
a=xlm['date'] > "2020-12-31 00:00:00"
xlm=xlm[a]
high=pd.DataFrame(xlm,columns=['high'] )
low=pd.DataFrame(xlm,columns=['low'])
high=pd.DataFrame(high,range(0,101))
low=pd.DataFrame(low,range(0,101))
vusdt=xlm['Volume USDT']
vxlm=xlm['Volume XLM']
xlm
```

```python
p = np.array(high)
n = np.array(low)
arr = []
for i in range(0,100):
    arr.insert(i*2,p[i,0])
    arr.insert((i*2)+1,n[i,0])


arr=np.array(arr).reshape(100,2)
list1 = arr.tolist()
```

```python
from scipy.spatial import distance
#euclidean_distances-for hige and low in one days
print(distance.euclidean(arr[0], arr[5]))
#chebyshev
print(distance.chebyshev(arr[0], arr[5]))
#manhattan
print(distance.cityblock(arr[0], arr[5]))
#minkowski
print(distance.minkowski(arr[0], arr[5],2))
#cosine similarity
print(distance.cosine(arr[0],arr[5]))
```

```python
from math import sqrt
def euclidean_dist(doc1, doc2):

    pre_square_sum = 0
    for idx,_ in enumerate(doc1):
        pre_square_sum += (doc1[idx] - doc2[idx]) ** 2
    
    return sqrt(pre_square_sum)
```

```python
#euclidean_distances
matrix=np.zeros([10,10],float)
for i in range(0,10):
    for j in range(i+1):
       matrix[i,j]=euclidean_dist(list1[i],list1[j])
matrix       
```

```python
#euclidean_distances_chart
plt.imshow(matrix, interpolation="nearest", origin="upper")
plt.colorbar()
plt.show()
```
