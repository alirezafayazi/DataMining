```python
#average
Sum=sum(high)
average= Sum/len(high)
print (average)
```

```python
#middle
high=high.sort_values(by=['high'])
middle=high[(len(high)-1)//2:(len(high)+2)//2]
print(middle)
if len(high)%2 == 0:
    print((middle[0]+middle[1])/2)
    #middle with nympy
np.median(high)
```

```python
var=[]
#my code To calculate the variance
for i in range(0,305) :
    var.insert(i,(high[i]-average)**2)
Sum_var=sum(var)
Variance=Sum_var/len(var)
print(Variance)
#calculate the variance with numpy
print(np.var(high))
```

```python
#Standard deviation 
print('Standard deviation is : ')
np.std(high)
```

```python
Q1=xlm['high'].quantile(0.25)
Q2=xlm['high'].quantile()
Q3=xlm['high'].quantile(0.75)

print("Q1=",Q1,"Q2=",Q2,"Q3=",Q3)
iqr=Q3-Q1
print("IQR=" , iqr)
lower_bound = Q1 -(1.5 * iqr) 
upper_bound = Q3 +(1.5 * iqr) 
print("upper bound is : ",upper_bound , "  _   lower bound is :", lower_bound)

#outliers
outliers=xlm[(xlm['high'] > 0.622665) | (xlm['high'] < 0.12126499999999998)]
print(outliers)
```
